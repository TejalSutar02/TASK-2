```python
import pandas as pd
```

```python
df=pd.read_csv('employee_data.csv')
```

```python
df.columns
```

```python
df
```

```python
#1.Rename Columns:
d={'full_name':'name',
  'emp_id':'employee_id',
  'emp_status':'employment_id'}
df.rename(columns=d,inplace=True)
df
```

```python
#2. Rearrange Columns:
df=df[df.columns[[0,1,2,3,4,6,7,5,8]]]
df
```

```python
#3. Update Column Values(Mapping):
d={'male':'Male','female':'Female'}
df['gender']=df['gender'].map(d)
df
```

```python
#4.Remove Unwanted Characters
df['name']=df['name'].str.strip()
df
```

```python
#7. Handle Missing Values:
#Fill missing values in the age column with the average age of the employees.
df['age']=df['age'].fillna(df['age'].mean())
df
```

```python
# Fill missing values in the salary column with the median salary
df['salary']=df['salary'].fillna(df['salary'].median())
df
```

```python
# For missing join_date, use forward fill (ffill).
df['join_date']=df['join_date'].fillna(method='ffill')
df
```

```python
#5.Change Data Type:
# Convert the age column to integer type,
df['age']=df['age'].astype(int)
df
```

```python
# the salary column to float type.
df['salary']=df['salary'].astype(float)
df
```

```python
#6. Replace Specific Values:
# In the employment_status column, replace missing values with 'contract'.
df['employment_id']=df['employment_id'].fillna('contract')
df
```

```python
#8.Remove Duplicates:
# Ensure there are no duplicate rows in the dataset.
df.drop_duplicates(inplace=True)
df
```

```python
#9.Apply String Function:
# Capitalize the first letter of each word in the department column.
df['department']=df['department'].str.capitalize()
df
```

```python
#10.Drop Unwanted Columns:
# Remove the exp_years column from the dataset.
df.drop('exp_years',axis=1,inplace=True)
df
```

```python
#11. Remove Specific Rows:
# Drop rows where salary is above 80,000 to focus on mid-level employees
df.drop(df[df['salary']>80000].index)
```

```python
#12.Reset Index:
#After cleaning, reset the index of the DataFrame
df.reset_index(inplace=True)
df
```

```python
#13. Set Column as Index:
#Set the employee_id as the index of the DataFrame
df.set_index('employee_id',inplace=True)
df
```
